```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.linalg import pinv


################################## Task 1.1 ##########################################

def linear_regression(X, y):
    """Generalize the given training data to a linear function using linear regression
       as described in the assignment and return the function's weights.
       
       Arguments:
       X: The design matrix with a prepended column containing 1 for every row for bias
          simplification.
       y: The desired output for every entry (row) in the design matrix.
       
       Returns:
       A vector containing the learned weights of the function and as first entry the
       bias.
       
       """

    w = np.array(pinv(X.transpose() @ X) @ X.transpose() @ y)
    return w


################################## Task 1.2 ##########################################

def get_data_set(data_path, dim):
    """ Load the data set at the given location and prepare it for linear regression.
    
    Arguments:
    data_path: The path to the file containing the data. The file must be a .csv of the
               form described in the assignment.
    dim: The dimension of each example. E.g. The entry (x1, x2, y) has dimension 2.
    
    Returns:
    A pair containing as first component the design matrix of the loaded data with a
    prepended column containing 1 for each entry for bias simplification and as second
    component the corresponding output values as a vector.
    
    """
    
    """ load design matrix X and prepend 1 for bias simplification """
    X = pd.read_csv(data_path, usecols = range(0, dim))
    prefix_1 = np.array([[1] * X.shape[0]]).transpose()
    X = np.concatenate((prefix_1, X), axis=1)
    
    """ load output vector y """
    y = np.array(pd.read_csv(data_path, usecols = [dim]))
    
    return (X,y)

def mean_squared_error(X, y, w):
    """Calculate the mean squared error of a linaer function with the given weights used
       to predict the outcome for the given test data.
       
    Arguments:
    X: The design matrix of the test data.
    y: The desired output for every entry (row) in the design matrix.
    w: The weights of the linear function used to predict the outcome.
    
    Returns:
    The mean squared error.
    
    """
    
    error = X @ w - y
    squared_error = np.array(error.transpose() @ error)
    mean = 1/X.shape[0]
    return float(mean * squared_error)

train_set_2d = get_data_set('dataset/regression/train_2d_reg_data.csv', 2)
test_set_2d = get_data_set('dataset/regression/test_2d_reg_data.csv', 2)

bias_weights_2d = linear_regression(train_set_2d[0], train_set_2d[1])

print("Task 1.2:\n")
print("vector w containing bias and weights learned from the two dimensional",
      " training set: \n", bias_weights_2d, "\n\n")
print("mean squared error for the training set: ",
     mean_squared_error(train_set_2d[0], train_set_2d[1], bias_weights_2d),
     "\n")
print("mean squared error for the test set: ",
     mean_squared_error(test_set_2d[0], test_set_2d[1], bias_weights_2d),
     "\n")


################################## Task 1.3 ##########################################

train_set_1d = get_data_set('dataset/regression/train_1d_reg_data.csv', 1)
test_set_1d = get_data_set('dataset/regression/test_1d_reg_data.csv', 1)

bias_weight_1d = linear_regression(train_set_1d[0], train_set_1d[1])

print("\nTask 1.3:\n")

"""Calculate value paris for plotting the learned function."""
x_axis_vals = np.array(np.arange(0, 1.1, 0.1))
y_axis_vals = np.array([float(bias_weight_1d.transpose() @ np.array([[1], [val]]))
                        for val in x_axis_vals])

"""Plott learned function, training set and test set."""
learned = plt.plot(x_axis_vals, y_axis_vals, color = "C1", label = "learned function")
train_set = plt.scatter(train_set_1d[0][:,1], train_set_1d[1],
                        marker = ".", color = "C0", label = "training set entries")
test_set = plt.scatter(test_set_1d[0][:,1], test_set_1d[1],
                       marker = ".", color = "C2", label = "test set entries")
plt.legend()
plt.show()
```
